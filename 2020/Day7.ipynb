```python
length = 0
shiny_gold_holders = ['shiny gold']

while len(shiny_gold_holders) != length:
    length=len(shiny_gold_holders)
    for j in baggers:
        for k in shiny_gold_holders:
            if (k in j) and (j[:j.find("contain")-6] not in shiny_gold_holders):
                shiny_gold_holders.append(j[:j.find("contain")-6])
shiny_gold_holders = [i for i in shiny_gold_holders if i!='shiny gold']
sorted(shiny_gold_holders), len(shiny_gold_holders)
```
