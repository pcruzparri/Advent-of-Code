```python
f = open('Day4_input.txt', 'r').readlines()
card_data = {i.split(':')[0]:[j.split() for j in i.strip('\\n').split(':')[1].split('|')] for i in f}
card_data['Card   1']
```

### Soution 1

```python
sum([2**(sum([n in winning for n in nums])-1)*any([n in winning for n in nums])
     for card, (winning, nums) in card_data.items()])
```

### Solution 2

```python
import sys
all_cards = 0

def won_cards(card_num):
    global card_data, all_cards
    
    all_cards+=1
    winning, nums = card_data[f'Card{" "*(4-len(str(card_num)))}{card_num}']
    cards_won = sum([n in winning for n in nums])
    if cards_won:
        return [won_cards(n) for n in range(card_num+1, card_num+cards_won+1, 1)]
    else: 
        return 0



for card in card_data.keys():
    sys.stdout.flush()
    print(card, '/', len(card_data), sep='', flush=True)
    won_cards(int(card.split()[1]))
    
all_cards
```
