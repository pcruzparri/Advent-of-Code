```python
horver_lines = []
for i in points:
    if i[0]==i[2] and int(i[1])<int(i[3]):
        for j in range(int(i[1]),int(i[3])+1): 
            horver_lines.append((int(i[0]),j))
            
    if i[0]==i[2] and int(i[1])>int(i[3]):
        for j in range(int(i[3]),int(i[1])+1): 
            horver_lines.append((int(i[0]),j))
            
    if i[1]==i[3] and int(i[0])<int(i[2]):
        for j in range(int(i[0]),int(i[2])+1): 
            horver_lines.append((int(i[1]),j))
            
    if i[1]==i[3] and int(i[0])>int(i[2]):
        for j in range(int(i[2]),int(i[0])+1): 
            horver_lines.append((int(i[1]),j))
    else:
        continue
#print(horver_lines)        
morethantwo = 0
seen  = []
horver_sort = sorted(horver_lines)
for i,j in enumerate(horver_sort):
    if horver_sort.count(j)>1 and j not in seen:
        morethantwo+=1
        seen.append(j)

morethantwo
```
